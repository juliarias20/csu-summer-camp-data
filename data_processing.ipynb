```python
import pandas as pd
df = pd.read_csv('october_data.csv')

encoded_df= pd.get_dummies(df, columns=["Participated in an internship on campus", "Participated in an internship off campus", "Participated in undergraduate research", "Worked a job on campus", "Worked a job off campus"])
encoded_df.head()
```

```python
!pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select columns related to 'Participated in an internship on campus'
internship_on_columns = [col for col in encoded_df.columns if col.startswith("Participated in an internship on campus_")]
internship_on_df = encoded_df[internship_on_columns]

plt.figure(figsize=(6, 4))
sns.heatmap(internship_on_df.head(20), annot=True, cbar=True, cmap="YlGnBu")
plt.title("Heatmap of 'Participated in an internship on campus' Encodings (First 20 Rows)")
plt.xlabel("Encoding")
plt.ylabel("Sample Index")
plt.show()
```
